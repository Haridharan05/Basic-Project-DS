```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/complete.csv")
data.head(5)
```

```python
data.columns
```

```python
data.describe()
```

###Relating the variable with Scatterplots

```python
sns.relplot(x="New cases",y="New recovered", data=data)
```

```python
sns.pairplot(data)
```

```python
sns.relplot(x="Total Confirmed cases",y="New cases",kind='line',data=data)
```

```python
data.columns
```

```python
sns.catplot(x='Total Confirmed cases',y='New cases',data=data)
```

```python
sns.kdeplot(x = 'Total Confirmed cases' , data = data , color = 'black')
```

###Distribution plot

```python
sns.displot(x = 'Total Confirmed cases',kde=True,bins = 5 , data =data)
```

###Joint Plot

```python
sns.jointplot(x='Total Confirmed cases',y='New cases',data=data,kind='reg')
```

###Heatmaps

```python
corr = data.corr()
sns.heatmap(corr)
```

```python
sns.heatmap(corr,cmap=['red','green','blue'],
            vmin = 10,vmax=100000,annot=True)
```
